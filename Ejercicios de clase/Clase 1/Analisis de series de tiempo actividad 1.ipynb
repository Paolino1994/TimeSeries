# 3 - Redactar un informe describiendo 3 series de tiempo distintas

Serie de tiempo 1:
    La primera podria ser la cantidad de saldo en la sube, cada paso depende del anterior, siempre dismuinira con cada viaje que hagamos y subira en cuanto le realizemos una recarga, si bien tendria picos pronunciados en la subida, esto es propio de la naturaleza humana, donde uno siempre deja cada cosa que puede hacer para el final, en este caso, hasta que no se tengan $-50 de saldo, uno no suele recargar.

```python
pasajes=[25,30,35]
saldo = np.zeros(100)
saldo[0]=735
for i in range(99):
    rand = random.randint(0,2)
    if(saldo[i]<50):
        saldo[i+1]=saldo[i]+random.randint(100,700)
    else:
        saldo[i+1]=saldo[i]-pasajes[rand]
        
plt.plot(saldo)
plt.show()
```

Serie de tiempo 2:
    El peso de una persona, el peso es 100% dependiente de el peso que tuvimos en tiempos anteriores, es bastante mas cambiante que el la cantidad de saldo de la sube.


```python
peso = np.zeros(100)
peso[0]=70
for i in range(99):
    rand = random.randint(0,2)-1
    peso[i+1]=peso[i]+rand
        
plt.plot(peso)
plt.show()
```

Serie de tiempo 3:
