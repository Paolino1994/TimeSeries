```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
```

Lectura del Dataset

```python
df = pd.read_csv('Dataset/AEP_hourly.csv', sep=',')
```

```python
df
```

```python
fig, ax = plt.subplots(1,1,figsize=(26,5))
```

```python
plt.rcParams["figure.figsize"] = (20,5)
ax = df.plot(x="Datetime",y='AEP_MW',rot=90)
ax.set_ylabel("Consumo total en megawatts", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
ax.set_title("Consumo de electricidad de la proveedora de energia American Electric Power Company, Inc ")
```

```python
df["Datetime2"] = pd.to_datetime(df["Datetime"])
```

```python
df = df[df["Datetime2"]>=datetime.datetime(2015, 1, 1)]
df = df[df["Datetime2"]<=datetime.datetime(2018, 12, 31)]
```

```python
plt.rcParams["figure.figsize"] = (20,5)
ax = df.plot.scatter(x="Datetime",y='AEP_MW',rot=90)
ax.set_ylabel("Consumo total en megawatts", fontsize=16)
ax.set_xlabel("Fecha", fontsize=16)
ax.set_title("Consumo de electricidad de la proveedora de energia American Electric Power Company, Inc ")
```

```python
fig, ax = plt.subplots(4,1,figsize=(26,5))
for i in range(1):
    dfx = df[df["Datetime2"]>=datetime.datetime(2015+i, 1, 1)]
    dfx = dfx[dfx["Datetime2"]<=datetime.datetime(2015+i, 12, 31)]
    ax[i] = dfx.plot.scatter(x="Datetime2",y='AEP_MW',rot=90)
    ax[i].set_ylabel("Consumo total en megawatts", fontsize=16)
    ax[i].set_xlabel("Fecha", fontsize=16)
    ax[i].set_title("Consumo de electricidad del aÃ±o "+str(2015+i)+" de la proveedora de energia American Electric Power Company, Inc ")
    
```

```python
df["Date"]=df["Datetime2"].dt.date
```

```python
df2 = df.groupby(["Date"]).agg({'AEP_MW': [np.max]})
```

```python
df2["Date2"] = df2.index
```

```python
df2
```

```python
fig, ax = plt.subplots(4,1,figsize=(26,5))
for i in range(4):
    dfx = df2[df2["Date2"]>=datetime.date(2015+i, 1, 1)]
    dfx = dfx[dfx["Date2"]<=datetime.date(2015+i, 12, 31)]
    ax[i] = dfx.plot(x="Date2",y='AEP_MW',rot=90)
    ax[i].set_ylabel("Consumo total en megawatts", fontsize=16)
    ax[i].set_xlabel("Fecha", fontsize=16)
    ax[i].set_title("Consumo de electricidad del aÃ±o "+str(2015+i)+" de la proveedora de energia American Electric Power Company, Inc ")
    
```
